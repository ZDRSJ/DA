```python
# data 불러오기 
import pandas as pd

df = pd.read_csv('./dataset/crawling_data.csv')

df
```

```python
df = df.loc[:,['위치', '전용면적', '방/욕실수', '총세대수', '총동수', '교통시설', '학교시설', '편의시설']]

df
```

```python
# 특정 컬럼 전처리 및 분리
# 위치(동/번지) 분리
# df= df[['adr_gu', 'adr_dong', '번지', '본번', '부번', '도로명', '단지명', '전용면적(㎡)', '면적라벨', '거래금액(만원)', 'm2당금액(만원)', '건축년도', '계약년', '계약월']]
df['지번주소'] = df.위치.str.split('\n ').str[0]
df['adr_dong'] = df.지번주소.str.split(' ').str[2]
df['번지'] = df.지번주소.str.split(' ').str[3]

df
```

```python
# 전용면적
# df= df[['adr_gu', 'adr_dong', '번지', '본번', '부번', '도로명', '단지명', '전용면적(㎡)', '면적라벨', '거래금액(만원)', 'm2당금액(만원)', '건축년도', '계약년', '계약월']]
df['전용면적(㎡)'] = df.전용면적.str.split('㎡').str[0]

# 방/욕실수
df['room'] = df.방/욕실수.str.split('/').str[0] # 방개수
df['bath'] = df.방/욕실수.str.split('/').str[1] # 욕실수

# 총세대수
df[''] = df.방/욕실수.str.split('/').str[0]
```
