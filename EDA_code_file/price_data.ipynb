```python
# data 불러오기 
import pandas as pd

price_data = pd.read_csv('../dataset/price_data.csv')

price_data
```

```python
# 불필요한 열 삭제
df = price_data.iloc[:,2:]

df
```

```python
# 컬럼 재설정
df.columns = ["시군구", "번지", "본번", "부번", "단지명", "전용면적(㎡)", "계약년월", "계약일", "거래금액(만원)", "층", "건축년도", "도로명", "해제사유발생", "거래유형", "중개사소재지"]

df
```

```python
# 특정 컬럼 전처리 및 분리
# 시군구 분리
df['adr_gu'] = df.시군구.str.split(' ').str[1]
df['adr_gu'] = df.adr_gu.str.split('고양').str[1] # 고양 / 00구 분리
df['adr_dong'] = df.시군구.str.split(' ').str[2]

# 계약년월 분리
df['계약년월'] = df['계약년월'].astype(str)
df['year']  = df['계약년월'].map(lambda x : x[:4]) # 년
df['month']  = df['계약년월'].map(lambda x : x[4:]) # 월

df
```

```python
# 거래금액(만원) 컬럼 object -> int로 바꾸기
# "," 빼고 int로 바꾸는 함수
def to_numeric(string):
    return float(string.replace(',',''))

df['거래금액(만원)'] = df['거래금액(만원)'].apply(to_numeric)
df.dtypes #확인
```

```python
# 거래금액 ( 거래금액 /전용면적 → m2당 금액) 산출
df['m2당금액(만원)'] = df['거래금액(만원)']/df['전용면적(㎡)']
```

```python
'''
전용면적 (면적별로 분류하여 level별 라벨 인코딩
    - 60m2 이하 : 1
    - 60m2 ~ 83m2 : 2
    - 84m2 이상 : 3
'''

# 함수 만들기
def label(x):
    if x < 60:
        return 1
    elif (x >= 60) & (x < 84):
        return 2
    else:
        return 3


df['면적라벨'] = df['전용면적(㎡)'].apply(label)
```

```python
df
```

```python
# 불필요한 컬럼 제거
df.drop(['시군구', '계약년월', '계약일', '층', '해제사유발생', '거래유형', '중개사소재지'], axis= 1)

# 컬럼 재배치
df= df[['adr_gu', 'adr_dong', '번지', '본번', '부번', '도로명', '단지명', '전용면적(㎡)', '면적라벨', '거래금액(만원)', 'm2당금액(만원)', '건축년도', 'year', 'month']]

df
```
