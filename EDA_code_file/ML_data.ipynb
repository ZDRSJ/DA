### ML data
머신러닝 돌리기 전 종속 변수 제거 등 작업

```python
# data 불러오기
import pandas as pd

df_ml = pd.read_csv('../table/csv_general/total_merge_data_eda.csv')
```

```python
df_ml.drop('Unnamed: 0', axis=1, inplace= True)
```

```python
'''
종속되는 컬럼인 주소 동만 남기고 drop
종속되는 컬럼인 면적 하나 drop
종속되는 컬럼인 금액 하나 drop
'''
df_ml.drop(['주소', '면적라벨', 'm2당금액(만원)'], axis = 1, inplace = True)
```

```python
df_ml
```

```python
df_ml.info()
```

```python
# 방, 욕실 개수 int로 타입 변경
df_ml['room'] = df_ml['room'].apply(int)
df_ml['bath'] = df_ml['bath'].apply(int)
```

```python
# target 컬럼인 가격 컬럼 쉽게 바꾸기
df_ml.rename(columns = {'거래금액(만원)':'price'}, inplace=True)
```
