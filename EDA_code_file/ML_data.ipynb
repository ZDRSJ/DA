```python
# 단지명 : Frequency Encoding
fe = df_ml.groupby("단지명").size()/len(df_ml)
df_ml.loc[:, "단지명_enc"] = df_ml["단지명"].map(fe)
df_ml
```

```python
dong_list = list(df_ml.adr_dong.unique())
dong_list
```

```python
# adr_dong : Pandas의 get_dummies를 활용
df_ml = pd.get_dummies(df_ml, prefix=["enc"], columns=['adr_dong'])
df_ml
```

```python
enc_list = ['enc_가좌동', 'enc_고양동', 'enc_관산동', 'enc_대화동',
        'enc_덕은동', 'enc_덕이동', 'enc_도내동', 'enc_동산동', 'enc_마두동', 'enc_백석동',
        'enc_사리현동', 'enc_삼송동', 'enc_성사동', 'enc_성석동', 'enc_식사동', 'enc_신원동',
        'enc_원흥동', 'enc_일산동', 'enc_장항동', 'enc_주교동', 'enc_주엽동', 'enc_중산동',
        'enc_지축동', 'enc_탄현동', 'enc_토당동', 'enc_풍동', 'enc_행신동', 'enc_향동동',
        'enc_화정동']

print(len(enc_list))
print(len(dong_list))

enc_dict = {}
for enc in enc_list:
    for dong in dong_list:
        if dong in enc:
            enc_dict[enc] = dong

print(enc_dict)
print(len(enc_list))

# df_ml.rename(columns = enc_dict)
df_ml.rename(columns = enc_dict, inplace=True)
df_ml
```
