```python
# data 불러오기 
import pandas as pd

df = pd.read_csv('./dataset/crawling_data.csv')

df
```

```python
# 사용 컬럼만 불러오기
df = df.loc[:,['위치', '전용면적', '방/욕실수', '총세대수', '총동수', '교통시설', '학교시설', '편의시설']]

df
```

```python
# 컬럼별 결측값 확인
pd.set_option('display.max_rows', None)
df.isnull().sum()
```

```python
# 특정 컬럼 전처리 및 분리
# 위치(동/번지) 분리
df['지번주소'] = df.위치.str.split('\n ').str[0]
df['adr_dong'] = df.지번주소.str.split(' ').str[2]
df['번지'] = df.지번주소.str.split(' ').str[3]

# 전용면적
df['전용면적(㎡)'] = df.전용면적.str.split('㎡').str[0]

# 방/욕실수
df = df.rename(columns={'방/욕실수': '방욕실수'})
df['room'] = df.방욕실수.str.split('/').str[0] # 방개수
df['bath'] = df.방욕실수.str.split('/').str[1] # 욕실수
df['bath'] = df.bath.str.split('개').str[0] # 욕실수

# 총세대수
df['총세대수'] = df.총세대수.str.split('세대').str[0]

# 총동수
df['총동수'] = df.총동수.str.split('개동').str[0]

# 교통시설 라벨링 함수 정의
def traffic_labeling(x):
    if type(x) == float:  # 결측치의 타입이 float여서 해당 조건문 정의 - 교통시설 없으면 라벨 0(가장 하위)
        return 0
    elif '도보' in x:  # 도보, 도보&차량이면 라벨 2(가장 상위)
        return 2
    elif '차량' in x:  # 차량이면 라벨1(중간)
        return 1

df['traffic_label'] = df['교통시설'].apply(traffic_labeling)

# 학교시설 및 편의시설
pd.set_option('display.max_rows', 20)
df['학교시설'] = df.학교시설.str.split(',')
df['편의시설'] = df.편의시설.str.split(',')

# null == float 
df['학교시설'] = df['학교시설'].map(lambda x : 0 if type(x) == float else len(x))
df['편의시설'] = df['편의시설'].map(lambda x : 0 if type(x) == float else len(x))

df
```

```python
# 불필요한 컬럼 제거
df.drop(['위치', '전용면적', '방욕실수', '교통시설'], axis= 1)

# 컬럼 재배치
df= df[['지번주소', 'adr_dong', '번지', '전용면적(㎡)', '총세대수', '총동수', 'room', 'bath', 'traffic_label', '학교시설', '편의시설']]

df
```

```python
# 컬럼 리스트 정리
col_name = list(df.columns)
print(col_name)
print(type(col_name))
```

```python
# csv로 저장
df.to_csv("crawling_data_eda.csv",  # 경로
            sep = ",",  # 구분자
            columns =  col_name,  # 컬럼 이름 지정
            index = True)  # 행의 이름에 index 도 같이 내보냄
```
